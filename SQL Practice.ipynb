### Part 1: Basic Queries

1.1 List of albums, released after 2015, more than 10 tracks
Returns the title, release_year, and num_tracks for those albums, ordered by release_year from newest to oldest.

```python
print_sql("""
SELECT title, release_year, num_tracks
FROM albums
WHERE release_year > 2015
    AND num_tracks > 10
ORDER BY release_year DESC;
""")
```

### Part 2: Joins

 2.1 Write a query that joins the artists and albums tables and returns every
combination of artist name, album title, and release_year where the artist has
at least one album. Artists with no albums must not appear in the result.

```python
print_sql("""
SELECT artists.name, albums.title, albums.release_year
FROM artists
INNER JOIN albums
ON artists.id = albums.artist_id;
""")
```

2.2 One artist in the database has no albums linked to any stream records. Write a
query that uses LEFT JOIN across all three tables in this order: artists, then
albums, then streams. Return the artist name and stream_count for every artist,
including those with no matching streams. Then add a WHERE clause so your result
shows only the row where stream_count is NULL.

```python
print_sql("""
SELECT artists.name, streams.stream_count
FROM artists
LEFT JOIN albums
ON artists.id = albums.artist_id
LEFT JOIN streams
ON albums.id = streams.album_id
WHERE streams.stream_count IS NULL;
""")
```

2.3 SQLite does not support RIGHT JOIN natively. To retrieve all rows from the
streams table alongside matching album information, including streams whose
album_id has no match in the albums table, you must swap the table order and
use LEFT JOIN

Write a query that returns all 28 rows from streams with the matching title and
release_year from albums wherever a match exists. Return exactly these four
columns: streams.id, streams.stream_count, albums.title, albums.release_year.

```python
print_sql("""
SELECT streams.id, streams.stream_count, albums.title, albums.release_year
FROM streams
LEFT JOIN albums
ON streams.album_id = albums.id
""")
```

### Part 3: CASE WHEN and CAST

3.1
Write a query that returns the title, num_tracks, and a new column called
length_label for every album in the albums table. Use CASE WHEN to assign the
label according to these exact rules:


- Fewer than 9 tracks: 'Short'
- 9 to 12 tracks, inclusive on both ends: 'Standard'
- More than 12 tracks: 'Long'
Do not filter any albums out. All 20 albums must appear in the result.

```python
print_sql("""
SELECT title, num_tracks,
    CASE
        WHEN num_tracks < 9 THEN 'Short'
        WHEN num_tracks BETWEEN 9 AND 12 THEN 'Standard'
        ELSE 'Long'
    END AS length_label
FROM albums;
""")
```

3.2

Write a query that joins albums and streams on their shared id and returns the
album title, release_year, and stream_count. Use CAST to convert stream_count
to a REAL value and name that column stream_count_real. Limit the result to
10 rows.


In a markdown cell directly below this query, explain in two to three sentences
why dividing two INTEGER values in SQL can produce a misleading result, and how
casting to REAL fixes that.

```python
print_sql("""
SELECT albums.title, albums.release_year, CAST(streams.stream_count AS REAL) AS stream_count_real
FROM albums
INNER JOIN streams
ON albums.id = streams.album_id
LIMIT 10;
""")
```

Dividing two Integer values in SQL can produce a misleading result because SQL often treats that operation as integer division and therefore will drop the decimal, which can produce a misleading or inaccurate value. Therefore, casting the values to real forces SQL to treat values as floating-point numbers and preserves the decimal portion of the number during division.

### Part 4: String Functions

4.1 Write a query that returns three columns for all 11 artists: name, genre (the
full original string), and primary_genre (only the text before the first comma).
Use SUBSTR and INSTR to extract the primary genre. Do not hardcode any genre
string.

```python
print_sql("""
SELECT name, genre, SUBSTR(genre, 1, INSTR(genre, ',')-1) AS primary_genre
FROM artists;
""")
```

4.2
Using the same SUBSTR and INSTR technique from Question 4.1, write a query that
returns the name, genre, and primary_genre for only the artists whose extracted
primary_genre is exactly 'Electronic'. Filter on the derived value, not on the
original genre column.

```python
print_sql("""
SELECT name, genre, SUBSTR(genre, 1, INSTR(genre, ',')-1) AS primary_genre
FROM artists
WHERE primary_genre IS 'Electronic';
""")
```

### Part 5 Subqueries and CREATE TABLE AS

5.1
Write a query that returns the album title and stream_count for every row in
streams whose stream_count is strictly greater than the average stream_count
across all rows in the streams table. Calculate the average using a subquery
inside your WHERE clause. Do not hardcode any number.

```python
print_sql("""
SELECT albums.title, streams.stream_count
FROM streams
INNER JOIN albums
ON streams.album_id = albums.id
WHERE stream_count > (SELECT AVG(stream_count) FROM streams);
""")
```

5.2
Using CREATE TABLE AS, save the full result of your Question 5.1 query into a
new table called popular_albums. Begin your cell with DROP TABLE IF EXISTS
popular_albums so the cell can be re-run without error. After creating the table,
write a second SELECT query that reads from popular_albums and prints all its rows
using print_sql.


```python
conn.execute("""
DROP TABLE IF EXISTS popular_albums;
""")

conn.execute("""
CREATE TABLE popular_albums AS
    SELECT albums.title, streams.stream_count
    FROM streams
    INNER JOIN albums
    ON streams.album_id = albums.id
    WHERE stream_count > (SELECT AVG(stream_count) FROM streams);
""")

print_sql("""
SELECT * FROM popular_albums;
""")
```

6.2

Write a single query that performs a cluster sample from the streams table using
platform as the cluster variable. The query must first randomly select exactly 2
distinct platforms using a subquery, then return all rows from streams that belong
to those 2 platforms. Do not hardcode any platform name anywhere in the query.

```python
print_sql("""
    SELECT * FROM streams
    WHERE platform IN (
        SELECT DISTINCT platform
        FROM streams
        ORDER BY RANDOM()
        LIMIT 2
    );
""")
```
